```python
import requests
import json
import pandas as pd
```

```python
r =  requests.get('https://data.covidactnow.org/latest/us/counties.OBSERVED_INTERVENTION.timeseries.json')
r
```

```python
data = json.loads(json.dumps(r.json()))
```

```python
print(data[0]['actualsTimeseries'][0]['date'])
```

```python
for i in range(len(data[0]['actualsTimeseries'])):
    print(data[0]['actualsTimeseries'][i]['date'])
    for j in range(len(data)):
        print(data[j]['fips'])
        print(data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases'])
        print(data[j]['actualsTimeseries'][i]['population'])
        print("\n")
```

```python
result = {}
for i in range(12):
    date = data[0]['actualsTimeseries'][i]['date']
    result[date] = {}
    for j in range(len(data)):
        fips = data[j]['fips']
        result[date][fips] = {}
        result[date][fips]['cumulativeConfirmedCases'] = data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases']
        result[date][fips]['population'] = data[j]['actualsTimeseries'][i]['population']

print(result)
```

```python
with open("sample.json", "w") as outfile: 
    json.dump(result, outfile)
```
