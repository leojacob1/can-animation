```python
import requests
import json
import pandas as pd
from datetime import date
```

```python
r =  requests.get('https://data.covidactnow.org/latest/us/counties.OBSERVED_INTERVENTION.timeseries.json')
r
```

```python
data = json.loads(json.dumps(r.json()))
```

```python
print(data[0])
```

```python
aprilUpdate = data[0]['actualsTimeseries'][70]['date']
aprilUpdate
```

```python
latestUpdate = data[0]['actualsTimeseries'][len(data[0]['actualsTimeseries'])-1]['date']
latestUpdate
```

```python
days = len(data[0]['actualsTimeseries'])
days
```

```python
result2 = {}
for j in range(len(data)):
    fips = data[j]['fips']
    totalCount = 0
    TodayCount = data[j]['actualsTimeseries'][204]['cumulativeConfirmedCases']
    YesterdayCount = data[j]['actualsTimeseries'][203]['cumulativeConfirmedCases']
    if TodayCount == None:
        TodayCount = 0
    if YesterdayCount == None:
        YesterdayCount = 0
    totalCount = TodayCount - YesterdayCount
    if totalCount >= 0:
        result2[fips] = totalCount  
    
    #result2[fips]= 'hello'
result2
```

```python
sort_orders2 = sorted(result2.items(), key=lambda x: x[1])
for i in sort_orders2:
    print(i[0], i[1])
```

```python
sort2 = {}
sort_orders2 = sorted(result2.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders2:
    fips = i[0]
    averageCaseDensity = i[1]
    sort2[fips] = averageCaseDensity
sort2
```

```python
today = date.today()
string = str(today) + "corrected_average_case_density_ordered"
date = string + ".json"
with open(date, "w") as outfile: 
    json.dump(sort2, outfile)
```
