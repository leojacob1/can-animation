```python
import requests
import json
import pandas as pd
from datetime import date
```

```python
r =  requests.get('https://data.covidactnow.org/latest/us/counties.OBSERVED_INTERVENTION.timeseries.json')
r
```

```python
data = json.loads(json.dumps(r.json()))
```

```python
result = {}
for i in range(len(data[0]['actualsTimeseries'])-3):
    date = data[0]['actualsTimeseries'][i]['date']
    result[date] = {}
    for j in range(len(data)):
        fips = data[j]['fips']
        result[date][fips] = {}
        if data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases'] != None:
            result[date][fips]['caseDensity'] = (int(data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases']) / int(data[j]['actualsTimeseries'][i]['population'])) * 100000
        else:
            result[date][fips]['caseDensity'] = None
```

```python
today = date.today()
string = str(today)
date = string + ".json"
with open(date, "w") as outfile: 
    json.dump(result, outfile)
```

```python
result = {}
for j in range(len(data)):
    fips = data[j]['fips']
    totalCaseDensity = 0
    for i in range(70, len(data[0]['actualsTimeseries'])-3,1):
        if data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases'] != None:
            totalCaseDensity += (int(data[j]['actualsTimeseries'][i]['cumulativeConfirmedCases']) / int(data[j]['actualsTimeseries'][i]['population'])) * 100000
    averageCaseDensity = totalCaseDensity / (len(data[0]['actualsTimeseries'])-3)
    result[fips] = averageCaseDensity
result
```

```python
sort_orders = sorted(result.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders:
    print(i[0], i[1])
```

```python
sort = {}
sort_orders = sorted(result.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders:
    fips = i[0]
    averageCaseDensity = i[1]
    sort[fips] = averageCaseDensity
sort
```

```python
today = date.today()
string = str(today) + "average_case_density_ordered"
date = string + ".json"
with open(date, "w") as outfile: 
    json.dump(sort, outfile)
```

```python
result2 = {}
for j in range(len(data)):
    fips = data[j]['fips']
    casesSinceApril = 0
    if data[j]['actualsTimeseries'][len(data[0]['actualsTimeseries'])-4]['cumulativeConfirmedCases'] != None:
        if data[j]['actualsTimeseries'][70]['cumulativeConfirmedCases'] != None:
            casesSinceApril = int(data[j]['actualsTimeseries'][len(data[0]['actualsTimeseries'])-4]['cumulativeConfirmedCases']) - int(data[j]['actualsTimeseries'][70]['cumulativeConfirmedCases'])
        else:
            casesSinceApril = data[j]['actualsTimeseries'][len(data[0]['actualsTimeseries'])-4]['cumulativeConfirmedCases']
    caseDensity = casesSinceApril / data[j]['actualsTimeseries'][len(data[0]['actualsTimeseries'])-4]['population'] * 100000
    averageCaseDensity = caseDensity / ((len(data[0]['actualsTimeseries'])-4) - 70)
    result2[fips] = averageCaseDensity
result2
```

```python
sort_orders2 = sorted(result2.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders2:
    print(i[0], i[1])
```
